```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

label_file = open("data/classes.txt", "r")

sortedClassContents = np.loadtxt("data/classes.txt")
float_formatter = "{:.0f}".format
np.set_printoptions(formatter={'float_kind':float_formatter})
classes = list(sortedClassContents)

print(sortedClassContents)
print(len(sortedClassContents))
```

```python
model = tf.keras.models.load_model('model_600')
```

```python
def get5TopPredict(predict):
    predicts = predict[0].argsort()[-5:][::-1]
    for i in predicts:
        print(classes[i])
```

```python
def predict(first):
    predict = model.predict([[classes.index(first)]])
    return classes[np.argmax(predict[0])]
```

```python
      
def predictDense(first):
    predict = model.predict([[classes.index(first)]])
    get5TopPredict(predict)
    return sortedClassContents[np.argmax(predict[0])]
```
